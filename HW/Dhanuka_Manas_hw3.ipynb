## Case 1: P(left) = P(right) = P(rest) = 1/3

```python
#importing required packages 
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
#Parent class form which child class will inherit properties to move the random walker
class base_walker: 
    def __init__(self):
        pass
    def one_step(self):
        pass
    def steps(self,n=1):
        for i in range(n):
            self.one_step()
```

```python
#class that generates random walk with the given criteria 

class random_walker(base_walker):
    def __init__(self):
        '''init function to initialize variables used in the class'''
        # define stored variables, initial data
        self.x = 0
        self.walk = [self.x]
        
    def one_step(self):
        '''Function to define criteria for one step; probalities of going right, left or staying still all are
        equal to 1/3'''
        rand = random.random()
        if rand > (2/3):
            self.x += 1
        elif (1/3) < rand <(2/3):
            self.x = self.x
        elif rand < (1/3):
            self.x -= 1
        self.walk.append(self.x)
        
    def get_walk(self):
        '''function to return the history of the position of the walker'''
        return self.walk
```

```python
def simulate_walks(time, num_walks):
    '''Function to simulate walks. Function takes in length of time for walk and number of simulations and returns
    a list of final positions of all the simulated walks. '''
    fin_positions = []
    for i in range (num_walks):
        walker = random_walker()
        walker.steps(time)
        positions = walker.get_walk()
        fin_positions.append(positions[-1])
    return fin_positions
```

```python
#Function calls for simulating 10s ,100s and 1000s walks 
fin_pos_1000 = simulate_walks(1000,1000000)
fin_pos_100 = simulate_walks(100,1000000)
fin_pos_10 = simulate_walks(10,1000000)
```

```python
# plotting probability density function

#Setting up the plot details 
plt.rcParams['figure.figsize'] = [15, 10] 
plt.xlabel('Positions', fontweight='bold', fontsize=12)
plt.ylabel('Probability Density', fontweight='bold', fontsize=12)
plt.title('Probability Density Function for Case 1', fontweight='bold', fontsize=14)
plt.grid(linestyle='--')

#Plotting the values 
plt.hist(fin_pos_1000, density = True, color = 'blue',bins= 40, label = "t = 1000s", alpha = 0.6)
plt.hist(fin_pos_100, density = True,color = 'red', bins = 28, label = "t = 100s", alpha = 0.5)
plt.hist(fin_pos_10, density = True, color = 'yellow', bins = 20, label = "t = 10s", alpha = 0.5 )
plt.legend(fontsize=12, borderpad = 0.8)
plt.show()
```

```python
#Converting lists to numpy arrays
arr_1000 = np.array(fin_pos_1000)
arr_100 = np.array(fin_pos_100)
arr_10 = np.array(fin_pos_10)

#Setting plot deatils 
plt.rcParams['figure.figsize'] = [15, 10] 
plt.xlabel('Positions', fontweight='bold', fontsize=12)
plt.ylabel('Probability Density', fontweight='bold', fontsize=12)
plt.title('Cumulative Density Function for Case 1', fontweight='bold', fontsize=14)
plt.grid(linestyle='--')

#Plotting cumulative density function for 1000s walks
count_1000, bins_count_1000 = np.histogram(arr_1000, bins=50)
pdf_1000 = count_1000 / sum(count_1000)
cdf_1000 = np.cumsum(pdf_1000)
plt.plot(bins_count_1000[1:], cdf_1000, label="t = 1000s", color = 'blue')

#Plotting cumulative density function for 100s walks
count_100, bins_count_100 = np.histogram(arr_100, bins=35)
pdf_100 = count_100 / sum(count_100)
cdf_100 = np.cumsum(pdf_100)
plt.plot(bins_count_100[1:], cdf_100, label="t = 100s", color = 'red')

#Plotting cumulative density function for 10s walks
count_10, bins_count_10 = np.histogram(arr_10, bins=20)
pdf_10 = count_10 / sum(count_10)
cdf_10 = np.cumsum(pdf_10)
plt.plot(bins_count_10[1:], cdf_10, label="t = 10s", color = 'yellow')

plt.legend(fontsize=12, borderpad = 0.8, loc = 'lower right')
plt.show()
```

## Case 2: P(right) = P(rest) = 1/2 ; P(left) = 0

```python
class random_walker_2(base_walker):
    def __init__(self):
        '''init function to initialize variables used in the class'''
        self.x = 0
        self.walk = [self.x]
        
    def one_step(self):
        '''Function to define criteria for one step; probalities of going right or staying still = 1/2 while
        going left = 0 '''
        rand = random.random()
        if rand > (1/2):
            self.x += 1
        elif rand <(1/2):
            self.x = self.x 
        self.walk.append(self.x)
        
    def get_walk(self):
        '''function to return the history of the position of the walker'''
        return self.walk
```

```python
def simulate_walks_2(time, num_walks):
    '''Function to simulate walks. Function takes in length of time for walk and number of simulations and returns
    a list of final positions of all the simulated walks. '''
    fin_positions = []
    for i in range (num_walks):
        walker = random_walker_2()
        walker.steps(time)
        positions = walker.get_walk()
        fin_positions.append(positions[-1])
    return fin_positions
```
